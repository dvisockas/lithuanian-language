```python
import os
import re
```

```python
path = '/home/danielius/Downloads/GARSYNAS_LIEPA_SuPriedais_Danieliui/AnotacijosANSIirUNICODE/Anotacijos_UNICODE/'
files = os.listdir(path)

target = open('compiled.txt','w+')
for file in files:
    file = open(f'{path}{file}', 'rb').read().decode('utf-16')
    if len(file) and len(file.split(' '))  > 2:
        regexes = {
            r'\d|_[\w]+': '',
            '  ': ' ',
            r'^\s+': '',
            r'\w+$': ''
        }
        
        for regex, replace in regexes.items():
            file = re.sub(regex, replace, file)
        target.write(f'{file}\n')
    
target.close() 
    
```
